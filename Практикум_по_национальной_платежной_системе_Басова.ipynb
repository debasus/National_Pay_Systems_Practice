### 1. Национальная платежная система (НПС)

Сущность и определение:

НПС — совокупность институциональных и инфраструктурных механизмов, обеспечивающих передачу денежных требований, включая обязательства центрального банка и коммерческих банков.
Включает операторов платежных систем, операторов электронных денег, банковских агентов и другие субъекты.

Структура НПС:

Ключевые участники: Банк России (регулятор), коммерческие банки, операторы платежных систем, агенты.

Взаимодействие строится на принципах надежности, безопасности и прозрачности.
Принципы работы:

Оперативное и безошибочное проведение расчетов.

Правовые и технологические нормы, регулирующие деятельность.

Механизм проведения расчетов: Перевод денежных средств через платежные системы с использованием сетей, счетов клиентов и межбанковских расчетов.

### 2. Платежные инструменты и технологии

Традиционные платежные системы:
Наличные и безналичные банковские переводы.

Электронные платежные системы:
Интернет-платежи, мобильные платежи, электронные кошельки.
Включают реальные и виртуальные платежные карты.

Платежные технологии:
Персональный идентификатор для безопасности.
Мобильный и онлайн-эквайринг.

Методы снижения рисков: шифрование, двухфакторная аутентификация, токенизация.

### 3. Роль электронных денег в современной системе платежей

Отличия электронных денег:
Они представляют собой цифровую форму денег (не только средства перевода).
Отличаются от оптовых и розничных систем по объему, способу эмиссии и сферам применения.

Эмиссия электронных денег:
Выпускается специализированными денежными посредниками (банками или операторами электронных денег).

Модели выбора розничного средства платежа:
Электронные деньги конкурируют с наличными и другими платежными способами по удобству, стоимости и скорости.

Сравнение издержек:
Электронные платежи зачастую дешевле и быстрее, особенно для массовых мелких платежей.

Мобильные платежи:
Рост популярности благодаря удобству использования смартфонов и интернет-приложений.

## 3. Практическое задание: Моделирование выбора розничного средства платежа
Напишите функцию, которая помогает выбрать между электронными деньгами и традиционными платежами, учитывая издержки и скорость.
python
Пример простого сравнения издержек


```python
def choose_payment_method(cost_e_money, cost_cash, speed_e_money, speed_cash):
    score_e = (1 / cost_e_money) + speed_e_money
    score_c = (1 / cost_cash) + speed_cash
    if score_e > score_c:
        return 'Электронные деньги'
    else:
        return 'Наличные'

# Тест
print(choose_payment_method(0.01, 0.03, 0.9, 0.5)) # Ожидается 'Электронные деньги'
```

Комментарий: функция способствует пониманию компромиссов при выборе платежного средства.
