## –ë–∞–∑–æ–≤—ã–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å –∏ —Ç–∏–ø—ã

–ß–∏—Å–ª–∞ –∏ –æ–ø–µ—Ä–∞—Ü–∏–∏:‚Äã —Ü–µ–ª—ã–µ int, –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ float;
–±–∞–∑–æ–≤—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã: +, -, *, /, //, %, **;
–æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ round() (–±—É—Ö–≥–∞–ª—Ç–µ—Ä—Å–∫–æ–µ –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ).‚Äã

–ü—Ä–∏–º–µ—Ä:


```python
a = 5
b = 2

print(a + b)      # 7
print(a / b)      # 2.5
print(a // b)     # 2 (—Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ)
print(a % b)      # 1 (–æ—Å—Ç–∞—Ç–æ–∫)
print(a ** b)     # 25

print(round(2.5))   # 2
print(round(3.5))   # 4
```

–°—Ç—Ä–æ–∫–∏:‚Äã
—Å–æ–∑–¥–∞–Ω–∏–µ: –æ–¥–∏–Ω–∞—Ä–Ω—ã–µ/–¥–≤–æ–π–Ω—ã–µ –∫–∞–≤—ã—á–∫–∏, –º–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏;
–∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è –∏ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ: +, *;
–≤–∞–∂–Ω—ã–µ –º–µ—Ç–æ–¥—ã: strip, lower, upper, replace, split, join, isnumeric, isalpha.‚Äã
–ü—Ä–∏–º–µ—Ä:


```python
s = "  Python is great!  "
print(s.strip())
print(s.lower())
print(s.replace("Python", "Data Science"))

words = s.strip().split()
print(words)
print(" | ".join(words))
print("2020".isnumeric(), "Hello".isalpha())
```

–ë—É–ª–µ–≤—ã –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è:‚Äã
—Ç–∏–ø bool: True –∏ False;
–æ–ø–µ—Ä–∞—Ç–æ—Ä—ã ==, !=, <, >, <=, >=;
–ª–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ and, or, not.‚Äã
–ü—Ä–∏–º–µ—Ä:


```python
age = 20
is_student = True

print(age >= 18 and is_student)
print(age < 18 or is_student)
print(not is_student)
```

–°—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö: —Å–ø–∏—Å–∫–∏, –∫–æ—Ä—Ç–µ–∂–∏, —Å–ª–æ–≤–∞—Ä–∏, –º–Ω–æ–∂–µ—Å—Ç–≤–∞
–°–ø–∏—Å–∫–∏:‚Äã
—Å–æ–∑–¥–∞–Ω–∏–µ, –∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è, —Å—Ä–µ–∑—ã;
–º–µ—Ç–æ–¥—ã: append, extend, insert, pop, remove, sort, reverse.‚Äã
–ü—Ä–∏–º–µ—Ä:


```python
nums = [10, 20, 30]
print(nums[0], nums[-1])
print(nums[1:])

nums.append(40)
nums.insert(1, 15)
print(nums)

nums.remove(30)
last = nums.pop()
print(nums, last)

nums.sort()
print(nums)
```

—É–¥–æ–±–Ω—ã –¥–ª—è —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å—Ç—Ä—É–∫—Ç—É—Ä (–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã, –∑–∞–ø–∏—Å–∏).‚Äã

```python
point = (10, 20)
x, y = point
print(x, y)
```

–°–ª–æ–≤–∞—Ä–∏:‚Äã
–ø–∞—Ä—ã ¬´–∫–ª—é—á‚Äì–∑–Ω–∞—á–µ–Ω–∏–µ¬ª;
–æ–ø–µ—Ä–∞—Ü–∏–∏: –æ–±—Ä–∞—â–µ–Ω–∏–µ –ø–æ –∫–ª—é—á—É, –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ, –º–µ—Ç–æ–¥—ã get, setdefault, –ø–µ—Ä–µ–±–æ—Ä –ø–æ items().‚Äã
–ü—Ä–∏–º–µ—Ä:


```python
person = {"name": "Anna", "age": 25}
print(person["name"])
print(person.get("city", "Unknown"))

person["city"] = "Moscow"
person.setdefault("job", "data scientist")

for key, value in person.items():
    print(key, "->", value)
```

–ú–Ω–æ–∂–µ—Å—Ç–≤–∞:‚Äã
—Ö—Ä–∞–Ω–µ–Ω–∏–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤;
–æ–ø–µ—Ä–∞—Ü–∏–∏: –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ, –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ, —Ä–∞–∑–Ω–æ—Å—Ç—å.‚Äã


```python
tags1 = {"python", "data", "ml"}
tags2 = {"data", "viz", "python"}

print(tags1 & tags2)  # –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ
print(tags1 | tags2)  # –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ
print(tags1 - tags2)  # —Ä–∞–∑–Ω–æ—Å—Ç—å
```

–£–ø—Ä–∞–≤–ª—è—é—â–∏–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏: if, for, while
–í–µ—Ç–≤–ª–µ–Ω–∏—è if/elif/else:‚Äã


```python
age = 15

if age >= 18:
    print("–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à—ë–Ω")
elif age >= 16:
    print("–ù—É–∂–µ–Ω —Å–æ–ø—Ä–æ–≤–æ–∂–¥–∞—é—â–∏–π")
else:
    print("–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â—ë–Ω")
```

–¶–∏–∫–ª for –∏ range:‚Äã


```python
for i in range(5):
    print("i =", i)

names = ["Anna", "Ivan", "Maria"]
for name in names:
    print("Hello,", name)

for i, name in enumerate(names, start=1):
    print(i, name)
```

–¶–∏–∫–ª while:‚Äã


```python
counter = 0
while counter < 3:
    print("–®–∞–≥", counter)
    counter += 1
```

List comprehension, map, filter:‚Äã


```python
nums = [1, 2, 3, 4, 5]
squares = [x**2 for x in nums if x % 2 == 0]
print(squares)

from math import sqrt
roots = list(map(sqrt, [9, 16, 25]))
print(roots)

text = "2020 - 1,87; 713 1,3"
digits = [ch for ch in text if ch.isnumeric()]
print(digits)
```

### –§—É–Ω–∫—Ü–∏–∏ –∏ –æ–±—Ä–∞–±–æ—Ç–∫–∞ –æ—à–∏–±–æ–∫

–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π:‚Äã
def, –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ;
–ø–æ–∑–∏—Ü–∏–æ–Ω–Ω—ã–µ –∏ –∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã, –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.‚Äã
–ü—Ä–∏–º–µ—Ä:


```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Anna"))
print(greet("Ivan", greeting="Hi"))
```

–ü—Ä–∏–º–µ—Ä –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:‚Äã


```python
def factorial(n: int) -> int:
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(factorial(5))  # 120
```

–ò—Å–∫–ª—é—á–µ–Ω–∏—è: try/except/else/finally:‚Äã

```python
from math import sqrt

x = input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: ")
try:
    value = float(x)
    result = sqrt(value)
except ValueError:
    print("–û—à–∏–±–∫–∞: –≤–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ")
else:
    print("–ö–æ—Ä–µ–Ω—å:", result)
finally:
    print("–ì–æ—Ç–æ–≤–æ")
```

# –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–∞–¥–∞–Ω–∏—è

–†–∞–±–æ—Ç–∞ —Å–æ —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ —Å–ø–∏—Å–∫–∞–º–∏:

–∑–∞–ø—Ä–æ—Å–∏—Ç—å —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å—Ç—Ä–æ–∫—É —Å –§–ò–û;
—Ä–∞–∑–±–∏—Ç—å –Ω–∞ —á–∞—Å—Ç–∏, –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ —Ñ–æ—Ä–º–∞—Ç—É ¬´–§–∞–º–∏–ª–∏—è –ò.–û.¬ª;
—Å–æ—Ö—Ä–∞–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –¥–ª—è –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ª—é–¥–µ–π –≤ —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π.

–°–ª–æ–≤–∞—Ä–∏ –∏ —á–∞—Å—Ç–æ—Ç—ã —Å–ª–æ–≤:‚Äã
–∑–∞–¥–∞—Ç—å —Ç–µ–∫—Å—Ç (–∫–æ—Ä–æ—Ç–∫–∏–π –∞–±–∑–∞—Ü);
–æ—á–∏—Å—Ç–∏—Ç—å –æ—Ç –ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏, –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É;
—Å –ø–æ–º–æ—â—å—é —Å–ª–æ–≤–∞—Ä—è –ø–æ—Å—á–∏—Ç–∞—Ç—å —á–∞—Å—Ç–æ—Ç—É —Å–ª–æ–≤ –∏ –≤—ã–≤–µ—Å—Ç–∏ —Ç–æ–ø-5.‚Äã

–ü—Ä–æ—Å—Ç—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ —É—Å–ª–æ–≤–∏—è:‚Äã
–Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é (¬´—Ä–µ–±—ë–Ω–æ–∫¬ª, ¬´–≤–∑—Ä–æ—Å–ª—ã–π¬ª, ¬´–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä¬ª);
–ø—Ä–∏–º–µ–Ω–∏—Ç—å –µ—ë –∫ —Å–ø–∏—Å–∫—É –≤–æ–∑—Ä–∞—Å—Ç–æ–≤ —á–µ—Ä–µ–∑ list comprehension;
–ø–æ—Å—á–∏—Ç–∞—Ç—å, —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ (–∏—Å–ø–æ–ª—å–∑—É—è —Å–ª–æ–≤–∞—Ä—å —Å—á—ë—Ç—á–∏–∫–æ–≤).


```python
people = [] # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –æ –ª—é–¥—è—Ö

print("–í–≤–µ–¥–∏—Ç–µ –§–ò–û (–§–∞–º–∏–ª–∏—è –ò–º—è –û—Ç—á–µ—Å—Ç–≤–æ). –û—Å—Ç–∞–≤—å—Ç–µ —Å—Ç—Ä–æ–∫—É –ø—É—Å—Ç–æ–π, —á—Ç–æ–±—ã –∑–∞–∫–æ–Ω—á–∏—Ç—å.")

while True: # –ó–∞–ø—É—Å–∫–∞–µ–º –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª –¥–ª—è –≤–≤–æ–¥–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –§–ò–û
    full_name_input = input("–§–ò–û: ").strip() # –ó–∞–ø—Ä–∞—à–∏–≤–∞–µ–º –§–ò–û –∏ —É–¥–∞–ª—è–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –ø–æ –∫—Ä–∞—è–º

    if not full_name_input: # –ï—Å–ª–∏ —Å—Ç—Ä–æ–∫–∞ –ø—É—Å—Ç–∞—è, –∑–Ω–∞—á–∏—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –∑–∞–∫–æ–Ω—á–∏–ª –≤–≤–æ–¥
        break # –í—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞

    parts = full_name_input.split() # –†–∞–∑–¥–µ–ª—è–µ–º –§–ò–û –Ω–∞ —á–∞—Å—Ç–∏ –ø–æ –ø—Ä–æ–±–µ–ª–∞–º
    formatted_name = "" # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–º–µ–Ω–∏

    if len(parts) >= 3: # –ï—Å–ª–∏ —á–∞—Å—Ç–µ–π —Ç—Ä–∏ –∏–ª–∏ –±–æ–ª—å—à–µ (–ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º –§–∞–º–∏–ª–∏—è –ò–º—è –û—Ç—á–µ—Å—Ç–≤–æ)
        surname = parts[0] # –ü–µ—Ä–≤–∞—è —á–∞—Å—Ç—å - —Ñ–∞–º–∏–ª–∏—è
        first_initial = parts[1][0].upper() if parts[1] else '' # –ü–µ—Ä–≤–∞—è –±—É–∫–≤–∞ –∏–º–µ–Ω–∏ –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ
        middle_initial = parts[2][0].upper() if parts[2] else '' # –ü–µ—Ä–≤–∞—è –±—É–∫–≤–∞ –æ—Ç—á–µ—Å—Ç–≤–∞ –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ

        formatted_name = f"{surname} {first_initial}.{middle_initial}." # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –∫–∞–∫ '–§–∞–º–∏–ª–∏—è –ò.–û.'
    elif len(parts) == 2: # –ï—Å–ª–∏ —á–∞—Å—Ç–µ–π –¥–≤–µ (–ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º –§–∞–º–∏–ª–∏—è –ò–º—è)
        surname = parts[0]
        first_initial = parts[1][0].upper() if parts[1] else ''
        formatted_name = f"{surname} {first_initial}." # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –∫–∞–∫ '–§–∞–º–∏–ª–∏—è –ò.'
    elif len(parts) == 1: # –ï—Å–ª–∏ —á–∞—Å—Ç—å –æ–¥–Ω–∞ (–ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º —Ç–æ–ª—å–∫–æ –§–∞–º–∏–ª–∏—è)
        formatted_name = parts[0] # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∫–∞–∫ –µ—Å—Ç—å
    else: # –ï—Å–ª–∏ —á—Ç–æ-—Ç–æ –Ω–µ–æ–∂–∏–¥–∞–Ω–Ω–æ–µ, —Å–æ—Ö—Ä–∞–Ω—è–µ–º –∏—Å—Ö–æ–¥–Ω–æ–µ –§–ò–û
        formatted_name = full_name_input

    person_data = { # –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞
        "original_fio": full_name_input, # –û—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–µ –§–ò–û
        "formatted_fio": formatted_name # –û—Ç—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–µ –§–ò–û
    }
    people.append(person_data) # –î–æ–±–∞–≤–ª—è–µ–º —Å–ª–æ–≤–∞—Ä—å –≤ —Å–ø–∏—Å–æ–∫ people

print("\n–†–µ–∑—É–ª—å—Ç–∞—Ç—ã:") # –í—ã–≤–æ–¥–∏–º –∑–∞–≥–æ–ª–æ–≤–æ–∫ –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
for p in people: # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É people
    print(p) # –í—ã–≤–æ–¥–∏–º –∫–∞–∂–¥—ã–π —Å–ª–æ–≤–∞—Ä—å (–¥–∞–Ω–Ω—ã–µ —á–µ–ª–æ–≤–µ–∫–∞)
```

```python
import string
from collections import Counter

# 1. –ó–∞–¥–∞–µ–º —Ç–µ–∫—Å—Ç (–∫–æ—Ä–æ—Ç–∫–∏–π –∞–±–∑–∞—Ü)
text = """–≠—Ç–æ –ø—Ä–∏–º–µ—Ä —Ç–µ–∫—Å—Ç–∞ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —á–∞—Å—Ç–æ—Ç—ã —Å–ª–æ–≤.
–¢–µ–∫—Å—Ç —Å–æ–¥–µ—Ä–∂–∏—Ç —Ä–∞–∑–ª–∏—á–Ω—ã–µ —Å–ª–æ–≤–∞ –∏ –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è.
–ú—ã —Ö–æ—Ç–∏–º –ø–æ—Å—á–∏—Ç–∞—Ç—å, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –∫–∞–∂–¥–æ–µ —Å–ª–æ–≤–æ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ —ç—Ç–æ–º —Ç–µ–∫—Å—Ç–µ.
–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–±–æ—Ç—É —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ –ø–µ—Ä–µ–Ω–æ—Å–∞–º–∏ —Å—Ç—Ä–æ–∫."""

# 2. –û—á–∏—â–∞–µ–º –æ—Ç –ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏ –∏ –ø—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
text_cleaned = text.lower()
text_cleaned = text_cleaned.translate(str.maketrans('', '', string.punctuation))

# –†–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞
words = text_cleaned.split()

# 3. –° –ø–æ–º–æ—â—å—é —Å–ª–æ–≤–∞—Ä—è —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É —Å–ª–æ–≤
word_counts = Counter(words)

# 4. –í—ã–≤–æ–¥–∏–º —Ç–æ–ø-5 –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è —Å–ª–æ–≤
print("–¢–æ–ø-5 –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è —Å–ª–æ–≤:")
for word, count in word_counts.most_common(5):
    print(f"'{word}': {count}")
```

```python
from collections import Counter

# 1. –§—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –≤–æ–∑—Ä–∞—Å—Ç—É –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é
def get_age_category(age: int) -> str:
    if age < 18:
        return "—Ä–µ–±—ë–Ω–æ–∫"
    elif 18 <= age < 65:
        return "–≤–∑—Ä–æ—Å–ª—ã–π"
    else:
        return "–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä"

# 2. –°–ø–∏—Å–æ–∫ –≤–æ–∑—Ä–∞—Å—Ç–æ–≤
ages = [5, 10, 25, 30, 70, 15, 45, 60, 80, 22]

# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ —Å–ø–∏—Å–∫—É –≤–æ–∑—Ä–∞—Å—Ç–æ–≤ —á–µ—Ä–µ–∑ list comprehension
age_categories = [get_age_category(age) for age in ages]
print("–ö–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤–æ–∑—Ä–∞—Å—Ç–æ–≤:", age_categories)

# 3. –ü–æ—Å—á–∏—Ç–∞—Ç—å, —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ (–∏—Å–ø–æ–ª—å–∑—É—è —Å–ª–æ–≤–∞—Ä—å —Å—á—ë—Ç—á–∏–∫–æ–≤)
category_counts = Counter(age_categories)

print("\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏:")
for category, count in category_counts.items():
    print(f"{category}: {count}")
```

–°–ª–æ–≤–∞—Ä–∏ –∏ dict comprehension:‚Äã

```python
text = "python data science python machine learning python"
words = text.split()
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)

# –ß–µ—Ä–µ–∑ dict comprehension
freq2 = {w: words.count(w) for w in set(words)}
print(freq2)
```

–†–∞–±–æ—Ç–∞ —Å –Ω–∞–±–æ—Ä–∞–º–∏ —Å–ª–æ–≤–∞—Ä–µ–π (—á–∞—Å—Ç—ã–π –ø–∞—Ç—Ç–µ—Ä–Ω –¥–∞–Ω–Ω—ã—Ö):‚Äã


```python
photos = [
    {"name": "photo1.jpg", "tags": {"coffee", "cup", "food"}},
    {"name": "photo2.jpg", "tags": {"food", "dinner", "vegetable"}},
    {"name": "photo3.jpg", "tags": {"city", "building"}},
]

# –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ —Ñ–æ—Ç–æ –ø–æ —Ç–µ–≥–∞–º
tag_index = {}
for photo in photos:
    for tag in photo["tags"]:
        tag_index.setdefault(tag, []).append(photo["name"])

for tag, names in tag_index.items():
    print(tag, "->", names)
```

–ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã (yield) –∏ –ª–µ–Ω–∏–≤—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è:‚Äã


```python
def even_numbers(limit):
    for x in range(limit + 1):
        if x % 2 == 0:
            yield x

for num in even_numbers(10):
    print(num, end=" ")
```

–∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—è try/except/else/finally –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ–≥–æ –≤–≤–æ–¥–∞/—á—Ç–µ–Ω–∏—è.‚Äã


```python
data = ["10", "abc", "5"]
numbers = []

for item in data:
    try:
        value = int(item)
    except ValueError:
        print(f"–ü—Ä–æ–ø—É—Å–∫–∞–µ–º –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {item}")
        continue
    numbers.append(value)

print("–ö–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ —á–∏—Å–ª–∞:", numbers)
```

–æ–±–µ—Å–ø–µ—á–∏—Ç—å, —á—Ç–æ–±—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–µ–≥–∞ —Å–ø–∏—Å–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π –±—ã–ª –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω –∏ –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–æ–≤.

```python
# 1. –°–æ–∑–¥–∞–µ–º —Ç–µ—Å—Ç–æ–≤—ã–π –ª–æ–≥-—Ñ–∞–π–ª
log_content = [
    "2025-01-01 INFO User logged in",
    "2025-01-01 WARNING Disk space low",
    "2025-01-02 ERROR Failed to connect to DB",
    "2025-01-02 INFO Data processed successfully",
    "2025-01-03 INFO Report generated",
    "2025-01-03 WARNING High CPU usage",
    "2025-04-01 ERROR API request failed",
    "2025-04-01 INFO System health check passed"
]

with open("log_file.txt", "w", encoding="utf-8") as f:
    for line in log_content:
        f.write(line + "\n")

print("–¢–µ—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª log_file.txt —Å–æ–∑–¥–∞–Ω.\n")

# 2. –°–∫—Ä–∏–ø—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –ª–æ–≥–æ–≤
log_counts = {
    'INFO': 0,
    'ERROR': 0,
    'WARNING': 0
}

try:
    with open("log_file.txt", "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if "INFO" in line:
                log_counts['INFO'] += 1
            elif "ERROR" in line:
                log_counts['ERROR'] += 1
            elif "WARNING" in line:
                log_counts['WARNING'] += 1
except FileNotFoundError:
    print("–û—à–∏–±–∫–∞: —Ñ–∞–π–ª log_file.txt –Ω–µ –Ω–∞–π–¥–µ–Ω.")
except Exception as e:
    print(f"–ü—Ä–æ–∏–∑–æ—à–ª–∞ –Ω–µ–ø—Ä–µ–¥–≤–∏–¥–µ–Ω–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–∏ —á—Ç–µ–Ω–∏–∏ —Ñ–∞–π–ª–∞: {e}")

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —É—Ä–æ–≤–Ω—è–º –ª–æ–≥–æ–≤:")
print(log_counts)
```

–ò—Å—Ç–æ—á–Ω–∏–∫: Python –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö, —Å—Ç—Ä. 250-251‚Äã

–õ–∏–Ω–µ–π–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ (Line plots) - —Å—Ç—Ä. 251-253:‚Äã


```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
x = np.linspace(0, 10, 100)
y = np.sin(x)

# –°–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–æ—Å—Ç–æ–≥–æ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)
plt.title('–°–∏–Ω—É—Å–æ–∏–¥–∞–ª—å–Ω–∞—è –≤–æ–ª–Ω–∞')
plt.xlabel('–û—Å—å X')
plt.ylabel('–û—Å—å Y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# –ù–µ—Å–∫–æ–ª—å–∫–æ –ª–∏–Ω–∏–π –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ
y2 = np.cos(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ sin(x) –∏ cos(x)')
plt.xlabel('–û—Å—å X')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar(label='–¶–≤–µ—Ç–æ–≤–∞—è —à–∫–∞–ª–∞')
plt.title('Scatter plot: –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è X –∏ Y')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ Y')
plt.show()

# Scatter plot —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏
categories = np.random.choice(['A', 'B', 'C'], 100)
plt.figure(figsize=(10, 6))
for category in ['A', 'B', 'C']:
    mask = categories == category
    plt.scatter(x[mask], y[mask], label=f'–ö–∞—Ç–µ–≥–æ—Ä–∏—è {category}', alpha=0.6)
plt.title('Scatter plot –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ Y')
plt.legend()
plt.show()
```

# Task
Create a Python script that generates a text file containing mixed data (numbers and text), reads it, extracts only numeric values (integers and floats), and then saves these cleaned numeric values into a new CSV file, with each number on a new line.

## –°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞

### Subtask:
–°–æ–∑–¥–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ —Ñ–∞–π–ª–∞ —Å –ø—Ä–∏–º–µ—Ä–∞–º–∏ —Å—Ç—Ä–æ–∫, —Å–æ–¥–µ—Ä–∂–∞—â–∏—Ö –∫–∞–∫ —á–∏—Å–ª–∞, —Ç–∞–∫ –∏ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π '–º—É—Å–æ—Ä–Ω—ã–π' —Ç–µ–∫—Å—Ç, –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª–∞ –æ—á–∏—Å—Ç–∫–∏.


**Reasoning**:
To begin the subtask, I need to create a list of mixed data (numbers and text) and then write these entries into a new text file named `mixed_data.txt`, ensuring each entry is on a new line.



```python
mixed_data = [
    "123",
    "45.67",
    "some text",
    "number 88.9 here",
    "just words",
    "-100",
    "12.34abc",
    "another line with 50 and text",
    "",
    "999.99",
    "data points: 10, 20, 30"
]

file_name = "mixed_data.txt"

with open(file_name, "w", encoding="utf-8") as f:
    for item in mixed_data:
        f.write(item + "\n")

print(f"–§–∞–π–ª '{file_name}' —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω —Å —Ç–µ—Å—Ç–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏.")
```

## –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

### Subtask:
–ü—Ä–æ—á–∏—Ç–∞—Ç—å —Å–æ–∑–¥–∞–Ω–Ω—ã–π —Ñ–∞–π–ª `mixed_data.txt`, –∏–∑–≤–ª–µ—á—å –∏–∑ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –≤—Å–µ —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (—Ü–µ–ª—ã–µ –∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ), –∏–≥–Ω–æ—Ä–∏—Ä—É—è —Ç–µ–∫—Å—Ç, –∏ —Å–æ–±—Ä–∞—Ç—å –∏—Ö –≤ —Å–ø–∏—Å–æ–∫.

#### Instructions
1. –û—Ç–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª `mixed_data.txt` –≤ —Ä–µ–∂–∏–º–µ —á—Ç–µ–Ω–∏—è ('r') —Å –∫–æ–¥–∏—Ä–æ–≤–∫–æ–π 'utf-8'.
2. –°–æ–∑–¥–∞–π—Ç–µ –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ `numeric_values` –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª.
3. –î–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –≤ —Ñ–∞–π–ª–µ:
    a. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ä–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è (–º–æ–¥—É–ª—å `re`) –∏–ª–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—é `str.replace()` –∏ `str.split()` –¥–ª—è –ø–æ–∏—Å–∫–∞ –∏ –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –≤—Å–µ—Ö —á–∏—Å–µ–ª. –£—á–∏—Ç—ã–≤–∞–π—Ç–µ –∫–∞–∫ —Ü–µ–ª—ã–µ, —Ç–∞–∫ –∏ –¥–µ—Å—è—Ç–∏—á–Ω—ã–µ —á–∏—Å–ª–∞ (—Å —Ç–æ—á–∫–æ–π –∏–ª–∏ –∑–∞–ø—è—Ç–æ–π).
    b. –ü—Ä–µ–æ–±—Ä–∞–∑—É–π—Ç–µ –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ –≤ `float`.
    c. –î–æ–±–∞–≤—å—Ç–µ –∫–∞–∂–¥–æ–µ —É—Å–ø–µ—à–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –≤ —Å–ø–∏—Å–æ–∫ `numeric_values`.
4. –û–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –æ—à–∏–±–∫–∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–æ–ª—å–∫–æ —Ç–µ–∫—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —á–∏—Å–ª–æ–º) —Å –ø–æ–º–æ—â—å—é `try-except` –±–ª–æ–∫–∞.
5. –ó–∞–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª.
6. –í—ã–≤–µ–¥–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ `numeric_values`.

**Reasoning**:
To extract numeric values from the `mixed_data.txt` file, I will read it line by line, use regular expressions to find all potential numbers, and then attempt to convert these into floats, handling any conversion errors.



```python
import re

numeric_values = []
file_name = "mixed_data.txt"

with open(file_name, "r", encoding="utf-8") as f:
    for line in f:
        # Find all numbers, including integers, floats, and negative numbers.
        # This regex looks for: optional '-' followed by one or more digits,
        # optionally followed by a '.' and one or more digits.
        numbers_in_line = re.findall(r"-?\d+(?:\.\d+)?", line)

        for num_str in numbers_in_line:
            try:
                value = float(num_str)
                numeric_values.append(value)
            except ValueError:
                # This handles cases where re.findall might pick up something
                # that looks like a number but isn't a valid float (e.g., just a dot).
                # However, with the current regex, this case is less likely.
                pass # Ignore values that cannot be converted to float

print("–ò–∑–≤–ª–µ—á–µ–Ω–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:", numeric_values)
```

## –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –æ—á–∏—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤ CSV

### Subtask:
–ó–∞–ø–∏—Å–∞—Ç—å –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –Ω–æ–≤—ã–π CSV-—Ñ–∞–π–ª, –≥–¥–µ –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ –±—É–¥–µ—Ç –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

#### Instructions
1. –û—Ç–∫—Ä–æ–π—Ç–µ –Ω–æ–≤—ã–π CSV-—Ñ–∞–π–ª —Å –∏–º–µ–Ω–µ–º `cleaned_numbers.csv` –≤ —Ä–µ–∂–∏–º–µ –∑–∞–ø–∏—Å–∏ ('w') —Å –∫–æ–¥–∏—Ä–æ–≤–∫–æ–π 'utf-8' –∏ `newline=''` –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã —Å `csv.writer`.
2. –°–æ–∑–¥–∞–π—Ç–µ –æ–±—ä–µ–∫—Ç `csv.writer`.
3. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ –≤ —Å–ø–∏—Å–∫–µ `numeric_values` –∑–∞–ø–∏—à–∏—Ç–µ –µ–≥–æ –≤ CSV-—Ñ–∞–π–ª, —É–±–µ–¥–∏–≤—à–∏—Å—å, —á—Ç–æ –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.
4. –ó–∞–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª.
5. –í—ã–≤–µ–¥–∏—Ç–µ —Å–æ–æ–±—â–µ–Ω–∏–µ, –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞—é—â–µ–µ —Å–æ–∑–¥–∞–Ω–∏–µ –∏ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ CSV-—Ñ–∞–π–ª–∞.

**Reasoning**:
The numeric values have been extracted and are available in the `numeric_values` list. Now, I need to save these values into a new CSV file, ensuring each number is on a separate line as specified in the subtask instructions.



```python
import csv

output_file_name = "cleaned_numbers.csv"

with open(output_file_name, "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    for num in numeric_values:
        writer.writerow([num])

print(f"–û—á–∏—â–µ–Ω–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —É—Å–ø–µ—à–Ω–æ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ —Ñ–∞–π–ª '{output_file_name}'.")
```
